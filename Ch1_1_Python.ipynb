1. Venn Diagrams and Events

Represent events using Python sets and compute unions, intersections, and complements.

```python
# Sample space
S = {1, 2, 3, 4, 5, 6}

# Events
A = {2, 4, 6}      # even numbers
B = {4, 5, 6}      # numbers greater than 3

# Union
A_union_B = A | B

# Intersection
A_intersection_B = A & B

# Complement of A
A_complement = S - A

print("A union B:", A_union_B)
print("A intersection B:", A_intersection_B)
print("Complement of A:", A_complement)
```

2. Discrete Probability Model
Goal

Assign probabilities to individual outcomes and sum themUntitled.jpg

```python
# Outcomes and probabilities
outcomes = ["R1", "R2", "B1", "B2", "B3"]

probabilities = {
    "R1": 1/5,
    "R2": 1/5,
    "B1": 1/5,
    "B2": 1/5,
    "B3": 1/5
}

# Event: selecting a red ball
event_A = ["R1", "R2"]

# Compute probability of event A
P_A = sum(probabilities[outcome] for outcome in event_A)

print("Probability of selecting a red ball:", P_A)
```

3. Equally Likely Outcomes

Compute probability using counting. 2.jpg

```python
# Sample space for a fair die
S = [1, 2, 3, 4, 5, 6]

# Event: number greater than 4
A = [x for x in S if x > 4]

# Probability
P_A = len(A) / len(S)

print("Event A:", A)
print("Probability of A:", P_A)
```

```python
# Number of cards
total_cards = 52
hearts = 13

P_heart = hearts / total_cards

print("Probability of drawing a heart:", P_heart)
```

# 4. Continuous Probability Models

Approximate continuous probability using random simulation.
3.jpg

```python
import random

# Simulate uniform random values between 0 and 10
n = 100000
count = 0

for _ in range(n):
    x = random.uniform(0, 10)
    if 3 <= x <= 7:
        count += 1

# Estimated probability
probability = count / n

print("Estimated probability:", probability)
```

# 5. Law of Total Probability

Compute overall probability from conditional probabilities.4.jpg

```python
# Given probabilities
P_A = 0.6      # Clinic A
P_B = 0.4      # Clinic B

P_D_given_A = 0.01
P_D_given_B = 0.03

# Law of total probability
P_D = P_D_given_A * P_A + P_D_given_B * P_B

print("Overall probability of disease:", P_D)
```
